```python
import pandas as pd
```

```python
df1 = pd.read_csv('1_500.csv')
df2 = pd.read_csv('3_1000.csv')
df3 = pd.read_csv('4_1000.csv')
df4 = pd.read_csv('5_500.csv')
#df5 = pd.read_csv('tiktok_feed_5.csv')
#df6 = pd.read_csv('tiktok_feed_6.csv')
combine = pd.concat([df1, df2, df3, df4])
combine
```

```python
combine['vid_id'].nunique()
```

```python
final = combine.drop_duplicates(subset=['vid_id'])
final
```

```python
final['user_name'].isnull().sum()
```

```python
final = final.dropna(subset=['user_name'])
final
```
