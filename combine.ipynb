```python
import pandas as pd
import numpy as np
```

# combine data

```python

hn4 = pd.read_csv('tiktok_feed.csv')
hn5 = pd.read_csv('tiktok_feed_combined.csv')
hn = pd.concat([hn4, hn5], ignore_index=True)
hn = hn[hn['vid_nview'].notna()]
hn = hn.drop_duplicates(subset=['vid_id'], keep='last')
hn = hn[hn['user_name'].notna()]
hn.reset_index(drop=True, inplace=True)
hn
```

```python
hn.to_csv('tiktok_feed_combined.csv', index=False)
```

# Extract most viral video

```python
data = pd.read_csv('finalProject\data\preprocessed_data.csv')
```

```python
top_videos = data.loc[data.groupby("user_name")["viral_score"].idxmax()]
top_videos = top_videos.sort_values(by="viral_score", ascending=False)
top_videos.to_csv("top_video_per_user.csv", index=False)
top_videos
```

```python
top_videos = pd.read_csv("raw_2.csv")

split_dfs = np.array_split(top_videos, 4)

for i, df_part in enumerate(split_dfs):
    df_part.to_csv(f"raw_2{i+1}.csv", index=False)
```

```python
top_videos = pd.read_csv("top_videos_part_1.csv")

split_dfs = np.array_split(top_videos, 4)

for i, df_part in enumerate(split_dfs):
    df_part.to_csv(f"part_{i+1}.csv", index=False)
```

```python
p1 = pd.read_csv('users_p1.csv')
p2 = pd.read_csv('users_p2.csv')
tp = pd.concat([p1, p2], ignore_index=True)
tp = tp[tp['vid_nview'].notna()]
tp = tp[tp['vid_duration'].notna()]
tp = tp.drop_duplicates(subset=['vid_id'], keep='last')
tp = tp[tp['user_name'].notna()]
tp.reset_index(drop=True, inplace=True)
tp
```

```python
tp.to_csv('tiktok_users.csv', index=False)
```
