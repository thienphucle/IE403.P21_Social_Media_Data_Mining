```python
import pandas as pd
```

```python
nm1 = pd.read_csv('tiktok_feed_nm.csv')
nm2 = pd.read_csv('tiktok_feed_nm2.csv')
nm = pd.concat([nm1, nm2], ignore_index=True)
nm = nm[nm['vid_nview'].notna()]
nm = nm.drop_duplicates(subset=['vid_id'], keep='last')
nm = nm[nm['user_name'].notna()]
nm.reset_index(drop=True, inplace=True)
nm.to_csv('tiktok_feed_nm.csv', index=False)
```

```python
hn = pd.read_csv('tiktok_feed_hn.csv')
hn1 = pd.read_csv('tiktok_feed_hn1.csv')
hn2 = pd.read_csv('tiktok_feed_hn2.csv')
hn3 = pd.read_csv('tiktok_feed_hn3.csv')
hn4 = pd.read_csv('tiktok_feed_hn4.csv')
hn5 = pd.read_csv('tiktok_feed_hn5.csv')
hn6 = pd.read_csv('tiktok_feed_hn6.csv')
hn7 = pd.read_csv('tiktok_feed_hn7.csv')
hn8 = pd.read_csv('tiktok_feed_hn8.csv')
hn = pd.concat([hn1, hn2, hn3, hn4, hn5, hn6, hn7, hn8], ignore_index=True)
hn = hn[hn['vid_nview'].notna()]
hn = hn.drop_duplicates(subset=['vid_id'], keep='last')
hn = hn[hn['user_name'].notna()]
hn.reset_index(drop=True, inplace=True)
hn.to_csv('tiktok_feed_hn.csv', index=False)
```

```python
df1 = pd.read_csv('tiktok_feed_combined2.csv')
df2 = pd.read_csv('tiktok_feed_tp.csv')
df3 = pd.read_csv('tiktok_feed_nm.csv')
df4 = pd.read_csv('tiktok_feed_hn.csv')
df5 = pd.read_csv('tiktok_feed_tpl.csv')
df6 = pd.read_csv('tiktok_feed_3.csv')
```

```python
df1 = df1[df1['vid_nview'].notna()]
df2 = df2[df2['vid_nview'].notna()]
df3 = df3[df3['vid_nview'].notna()]
df4 = df4[df4['vid_nview'].notna()]
df5 = df5[df5['vid_nview'].notna()]
df6 = df6[df6['vid_nview'].notna()]
combine = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)
combine = combine[combine['vid_nview'].notna()]
combine = combine.drop_duplicates(subset=['vid_id'], keep='last')
combine = combine[combine['user_name'].notna()]
combine.reset_index(drop=True, inplace=True)
combine
```
