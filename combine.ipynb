```python
import pandas as pd
```

```python

hn4 = pd.read_csv('tiktok_feed.csv')
hn5 = pd.read_csv('tiktok_feed_combined.csv')
hn = pd.concat([hn4, hn5], ignore_index=True)
hn = hn[hn['vid_nview'].notna()]
hn = hn.drop_duplicates(subset=['vid_id'], keep='last')
hn = hn[hn['user_name'].notna()]
hn.reset_index(drop=True, inplace=True)
hn
```

```python
hn.to_csv('tiktok_feed_combined.csv', index=False)
```
